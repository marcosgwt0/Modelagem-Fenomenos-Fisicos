# Função 1

$
f(x) = (x^2 - 4)(x^3 - 7x + 1)
$

Primeiro, calculamos a derivada da função $f(x)$:
\begin{align*}
f(x) &= (x^2-4)(x^3-7x+1) \\
f'(x) &= \frac{d}{dx} [(x^2-4)(x^3-7x+1)] \\
f'(x) &= \frac{d}{dx} (x^2-4)(x^3-7x+1) + (x^2-4)\frac{d}{dx} (x^3-7x+1) \\
f'(x) &= (2x)(x^3-7x+1) + (x^2-4)(3x^2-7) \\
f'(x) &= 5x^4 - 33x^2 + 2x + 28
\end{align*}

Substituímos $x=5$ para encontrar a inclinação da reta tangente no ponto $x=5$:
\begin{align*}
f'(5) &= 5(5)^4 - 33(5)^2 + 2(5) + 28 \
&= 2338\\
\end{align*}

O qual usando a função arco-tangente, temos:

\begin{align*}
\theta &= \arctan(m) \\
&= \arctan(2338) \\
&= 1.5703686108 \text{ radianos}.
\end{align*}

```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp
import math

#função 1
def f(x):
    return (x**2 - 4) * (x**3 - 7*x + 1)

x = np.linspace(-5, 5, 1000)
y = f(x)

plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('$f(x) = (x^2 - 4)(x^3 - 7x + 1)$')
plt.show()
```

# Função 2

$
f(x) = \sin(x^2)\cos(2x)
$

  Primeiramente, vamos calcular a derivada da função:

  \begin{align*}
  f(x) &= \sin(x^2)\cos(2x) \\
  f'(x) &= \cos(x^2) \cdot 2x \cdot \cos(2x) - \sin(x^2) \cdot \sin(2x) \cdot 2 \\
  f'(x) &= 2x\cos(x^2)\cos(2x) - 2\sin(x^2)\sin(2x)
  \end{align*}

  Agora, vamos calcular $f'(5)$:

  $f'(5) = 2(5)\cos \left(5^2\right)\cos \left(2(5)\right)-2\sin \left(2(5)\right)\sin \left(5^2\right)$
<br>
$f'(5) = 8.77861557$

  Assim, a inclinação da reta tangente em $x = 5$ é $ = f'(5) \approx 8.77861557$.

  O qual usando a função arco-tangente, temos:

\begin{align*}
\theta &= \arctan(m) \\
&= \arctan(8.77861557) \\
&= 1.45737207 \text{ radianos}.
\end{align*}

# Função 3

$
f(x) = x^2e^{-x}\ln(x)
$

  Derivada da função:
    \begin{align*}
    f(x) &= x^2 e^{-x} \ln(x) \\
    f'(x) &= \frac{d}{dx} (x^2 e^{-x} \ln(x)) \\
    f'(x) &= \frac{d}{dx} (x^2) e^{-x} \ln(x) + x^2 \frac{d}{dx} (e^{-x}) \ln(x) + x^2 e^{-x} \frac{d}{dx} (\ln(x)) \\
    f'(x) &= 2x e^{-x} \ln(x) - x e^{-x} \ln(x) + x^2 e^{-x} \frac{1}{x} \\
    f'(x) &= x e^{-x} (2 \ln(x) - 1)
    \end{align*}

  Inclinação da reta tangente no ponto $x=5$:
  \begin{align*}
  f'(5) &= 5 e^{-5} (2 \ln(5) - 1) \
  &\approx -0.12897487
  \end{align*}

  Portanto, a inclinação da reta tangente no ponto $x=5$ da função $f(x) = x^2 e^{-x} \ln(x)$ é aproximadamente $-0.12897487$.

  O qual usando a função arco-tangente, temos:

\begin{align*}
\theta &= \arctan(m) \\
&= \arctan(-0.12897487) \\
&= -0.128266779 \text{ radianos}.
\end{align*}

```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp
from sympy import *

#Função 3

def f(x):
    return x**2*np.exp(-x)*np.log(x)

x = np.linspace(0.1, 10, 1000)
y = f(x)

plt.plot(x, y)
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("Gráfico de f(x)=x^2e^{-x}ln(x)")
plt.show()
```

# Questão 2 (vale 2 pontos)
Uma empresa espacial está tentando descobrir o ponto máximo da órbita de um satélite cuja função orbital é dada por: 
$$f(x)=x^4−5x^3+10x^2−10x+5$$

Encontre o primeiro ponto de mínimo no intervalo $-10<=x<=10$.

Como sempre, você pode usar qualquer ferramenta disponível na internet para lhe auxiliar a resolver este problema. Lembre-se, contudo, que o sympy é uma boa alternativa. Em todos os casos, você precisará indicar o passo a passo da aplicação de todas as regras, técnicas, matrizes, ou tabelas que use para resolver este problema.

Em todas as questões você pode usar apenas os métodos, regras, técnicas e algoritmos que utilizamos na nossa disciplina, em sala, ou no material de apoio. Lembrem-se estamos limitados as bibliotecas sympy, numpy e matplotlib.

Para resolver esta questão você precisará de, no minimo, 4 casas decimais de precisão depois da vírgula. 

```python
import sympy as sp

# Cria o símbolo 'x' para usar em expressões simbólicas
x = sp.symbols('x')

# Define a função f(x) como x^4 - 5x^3 + (10x)^2 - 10x + 5
f = x**4 - 5*x**3 + (10*x)**2 - 10*x + 5

# Calcula a derivada de f(x) em relação a x
deriv = sp.diff(f, x)

# Inicializa a variável valorMin com o valor infinito
valorMin = float('inf')

# Itera sobre os valores de p de -10 a 10
for i in range(-10, 11):

    # Calcula a raiz da derivada de f(x) com valor de p
    y = sp.nsolve(deriv, i)

    # Verifica se a raiz está no intervalo [-10, 10]
    if -10 <= y <= 10:

        # Substitui x pela raiz encontrada na função f(x) e avalia o resultado
        valor = f.subs(x, y).evalf()

        # Verifica se o valor avaliado é menor do que o menor valor encontrado até agora
        if valor < valorMin:

            # Atualiza o menor valor encontrado e o ponto correspondente
            valorMin = valor
            pontoMin = y

# Calcula a segunda derivada de f(x) em relação a x
deriv2 = sp.diff(deriv, x)

# Verifica se a segunda derivada no ponto mínimo encontrado é positiva
if deriv2.subs(x, pontoMin) > 0:

    # Se sim, o ponto mínimo é exibido na tela
    print(f'{pontoMin} é um ponto de mínimo.')
else:

    # Se não, é exibido que o ponto não é de mínimo
    print(f'{pontoMin} não é um ponto de mínimo.')

# Exibe na tela o primeiro ponto de mínimo encontrado no intervalo [-10, 10] e seu valor mínimo
print(f'O primeiro ponto de mínimo da função no intervalo de [-10, 10] é: ({pontoMin}, {valorMin}).')
```

# Questão 3 (vale 2 pontos)
Considerando a função $f(x,y) = \sin(xy) + \cos(x+y)$, no intervalo $0\leq x\leq 5 \space\space e \space\space 0\leq y\leq 5$ trace, usando o plotly o gráfico desta função e marque os pontos de máximo e mínimo neste intevalo.

Lembre-se o sympy é sempre uma opção e você pode usar qualquer técnica, algoritmo, regra, ou tabela, que tenha sido usada na matéria. E que precisa de apenas uma casa decimal de precisão. Uma dica importante para a solução deste problema é a função **np.unravel_index**


```python
import numpy as np
import plotly.graph_objs as go

# Definindo a função
def f(x, y):
    return np.sin(x*y) + np.cos(x+y)

# Criando o grid de pontos
x = np.linspace(0, 5, 100)
y = np.linspace(0, 5, 100)
X, Y = np.meshgrid(x, y)

# Calculando os valores da função nos pontos do grid
Z = f(X, Y)

# Plotando a superfície da função
fig = go.Figure(data=[go.Surface(z=Z, x=X, y=Y)])
fig.update_layout(scene = dict(
                    xaxis_title='x',
                    yaxis_title='y',
                    zaxis_title='f(x,y)'),
                    margin=dict(l=0, r=0, t=0, b=0))
fig.show()

# Encontrando os pontos de máximo e mínimo
min_idx = np.unravel_index(np.argmin(Z), Z.shape)
max_idx = np.unravel_index(np.argmax(Z), Z.shape)
min_x, min_y, min_z = X[min_idx], Y[min_idx], Z[min_idx]
max_x, max_y, max_z = X[max_idx], Y[max_idx], Z[max_idx]

# Plotando os pontos de máximo e mínimo
fig.add_trace(go.Scatter3d(x=[min_x], y=[min_y], z=[min_z],
                           mode='markers', marker=dict(size=10, color='red')))
fig.add_trace(go.Scatter3d(x=[max_x], y=[max_y], z=[max_z],
                           mode='markers', marker=dict(size=10, color='red')))
fig.show()
```
